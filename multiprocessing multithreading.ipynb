**MULTIPROCESSING**

```python
import os
import time
import multiprocessing
import threading
```

```python
def india():
  print(f'CPU1 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'India: {i}')
def usa():
  print(f'CPU2 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'Usa: {i}')
if __name__=='__main__':
  print(f'main process id:{os.getpid()}')
  c1= multiprocessing.Process(target=india,args=())
  c2=multiprocessing.Process(target=usa,args=())

  c1.start()
  c2.start()

  c1.join()
  c2.join()
  print('main process completed')
```

**MULTITHREADING**

```python
def india():
  print(f'CPU1 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'India: {i}')
def usa():
  print(f'CPU2 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'Usa: {i}')
def uk():
  print(f'CPU3 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'Uk: {i}')
def canada():
  print(f'CPU4 process id:{os.getpid()}')
  for i in range(11,15):
    time.sleep(1)
    print(f'Canada: {i}')

def common1():
  t1=threading.Thread(target=india,args=())
  t2=threading.Thread(target=uk,args=())
  t1.start()
  t2.start()
def common2 ():
  t3=threading.Thread(target=usa,args=())
  t4=threading.Thread(target=canada,args=())
  t3.start()
  t4.start()

if __name__=='__main__':
  print(f'main process id:{os.getpid()}')
  c1= multiprocessing.Process(target=common1,args=())
  c2=multiprocessing.Process(target=common2,args=())

  c1.start()
  c2.start()

  c1.join()
  c2.join()
  print('main process completed')
```
