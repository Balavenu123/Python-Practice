```python
a=[[1, 2], [3, [4, 5]], [6]]
b=[]
for i in a:
  for j in i:
    if type(j)==list:
      for k in j:
        b.append(k)
    else:
      b.append(j)
print(b)
```

```python
def flatten_generator(nested_list):
    for item in nested_list:
        if isinstance(item, list):
            yield from flatten_generator(item)
        else:
            yield item
a=[[1, 2], [3, [4, 5,[4,5,6,7,[9,0,8,[8,1,[7,4,6,[0]]]]]]], [6]]
b = list(flatten_generator(a))
print(b)
```

```python
def check_list(a):
  new_ls = []
  for i in a:
    if type(i) == int:
      new_ls.append(i)
    else:
      new_ls.extend(check_list(i))
  return new_ls
out = check_list(a)
print(out)
```
