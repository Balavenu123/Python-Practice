```python
import numpy as np
```

```python
dir(np)
len(dir(np))
```

```python
# array is used to create 1D,2D,3D,4D,... data
# shape shows how many roows and columns
# size gives the no of values in the data
# ndim is used to find the dimention of the data
a=np.array([10,20,30,40,50,60])
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
```

```python
a=np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
# arange is used to generate only 1D data
b=np.arange(11,19)
print(b)
```

```python
# reshape is used to convert 1D to 2D,3D,....
b=np.arange(11,19)
print(b)
print('-------------------------')
c=b.reshape(2,4)
print(c)
```

```python
# reshape is use to conver from 1D to 2D only
# (-1,1) is one column
# (1,-1) is one row
a=np.array([10,11,12,13,14,15])
print(a)
print('--------------------')
b=a.reshape(-1,1)
print(b)
print('--------------------')
c=a.reshape(1,-1)
print(c)
```

```python
# zeros gives the o's data --> it meand black image
a=np.zeros((4,3),dtype=np.int32)
print(a)
```

```python
# ones is used to give 1's data
a=np.ones((4,3),dtype=np.int64)
print(a)
```

```python
#full is used to get data with our custom value
a=np.full((4,3),5)
print(a)
```

```python
# eye is used to get diagonal 1's and remaining 0's
a=np.eye(3,dtype=int,k=1)
print(a)
```

```python
# ravel is used to convert any dimentional into 1D

a=np.array([[1,2,3,4],[1,5,6,7],[9,0,8,7]])
print(a)
print('---------------')
b=np.ravel(a)
print(b)
```

```python
# Transpose is used to convert rows to column and column to row

a=np.array([[1,2,3,4],[1,5,6,7],[9,0,8,7]])
print(a)
print('--------------')
b=np.transpose(a)
print(b)
```

```python
# Concatination is used to merging the data
a=np.array([[10,11],[12,13]])
b=np.array([[100,200],[300,400]])
c=np.concatenate((a,b))
print(c)
```

```python
# randint is used to generate an integer randomly
a=np.random.randint(10,50,3)
print(a)
```

```python
# randn it will generate data randomly b/w -3 to 3 by just taking no of col & rows
a=np.random.randn(5,4)
print(a)
```

```python
#rand it will generate data randomly b/w 0 to 1 by just taking no of col & rows
a=np.random.rand(5,4)
print(a)
```

```python
 # axis=1 it is a row
 # axis=0 it is a column
a=np.array([[1,2,3,4],[1,5,6,7],[9,0,8,7]])
print(a)
print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.sum(a,axis=1))
print(np.min(a,axis=1))
print(np.max(a,axis=1))
```

```python
# argmax,argmin gives the index insted of value
a=np.array([[1,2,3,4],[1,5,6,7],[9,0,8,7]])
print(a)
print(np.argmin(a,axis=1))
print(np.argmax(a,axis=1))
```

```python
# mat operations
a=np.array([[10,11],[12,13]])
b=np.array([[100,200],[300,400]])
print(np.add(a,b))
print(np.subtract(a,b))
print(np.matmul(a,b))
print(np.mod(a,b))
```

```python
# repeat to repeat the given values either in row or column
a=np.arange(11,16)
b=a.reshape(1,-1)
c=np.repeat(b,3,axis=0)
print(c)
print('------------------')
b=a.reshape(-1,1)
c=np.repeat(b,2,axis=1)
print(c)
```

```python
# linespace is used to create data i.e evenly seperaes values
a=np.linspace(10,30,9,dtype=int)
print(a)
```

```python
# range is used to maintail the spacelimit
print(list(range(1,20,2)))
```
