```python
# Develop a function that uses a while loop to find the common elements between two lists.

a=[1,2,3,4,5]
b=[2,8,9,1,4]
e=[]
c=0
while c<len(a):
  ele1=a[c]
  d=0
  while d<len(b):
    ele2=b[d]
    if ele1==ele2:
      e.append(ele2)
    d=d+1
  c=c+1
print(e)
```
