```python
#Number Swapping without using 3rd variable
#model_1
a=10
b=20
a,b=b,a
print(f'a = {a} b = {b}')
```

```python
#model_2
a=10
b=20
a=a+b
b=a-b
a=a-b
print(f'a = {a} b = {b}')
```

```python
#Number Swapping using 3rd variable
a=10
b=20
c=a
a=b
b=c
print(f'a = {a} b = {b}')
```

```python
# Finding factorial of individual numbers in a number and their adding
import math
a=456
b=a%10   # b=6
c=a//10
#c=int(a/10)
d=c%10   # d=5
e=c//10  # e=4
#e=int(c/10)
fact_b=math.factorial(b)
fact_d=math.factorial(d)
fact_e=math.factorial(e)
print(fact_b+fact_d+fact_e)
```
