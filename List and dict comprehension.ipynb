```python
# Write a list comprehension to create a list of squares for numbers from 1 to 10.
print([i**2 for i in range(1,10)])
```

```python
# Generate a list of even numbers between 1 and 50 using list comprehension.
print([f'Even:{i}' for i in range(1,10) if i%2==0])
```

```python
# Convert all strings in a list to uppercase using list comprehens
print([i.upper() for i in ['bala','venu']])
```

```python
# Extract all vowels from a given string using list comprehension.
print([i for i in 'balavenubalineni' if i in 'aeiouAEIOU'])
```

```python
# Create a list of numbers that are divisible by both 2 and 3 from 1 to 30.
print([i for i in range(1,30) if i%2==0 and i%3==0])
```

```python
# Given a list of words, create a list of words that have more than 3 letters.
print([i for i in ['sun', 'moon', 'sky', 'stars'] if len(i)>3])
```

```python
# Flatten a 2D list using list comprehension.
print([j for i in [[1, 2], [3, 4], [5, 6]] for j in i])
```

```python
# Create a list of tuples (x, x**2) for all x from 1 to 10.
print([(i,i**2) for i in range(1,10)])
```

```python
# Replace all negative numbers in a list with 0 using list comprehension.
# [1, -3, 5, -2, 7] → [1, 0, 5, 0, 7]
print([i if i>0 else 0 for i in [1,-3,5,-2,7]])
```

```python
# Generate a list of all possible pairs (x, y) where x is from [1, 2, 3] and y is from [4, 5, 6].
print([(i,j) for i in [1,2,3] for j in (4,5,6)])
```

```python
# Create a list of prime numbers between 1 and 100 using list comprehension.
print([x for x in range(1,100) if x>1 if all(x%y!=0 for y in range(2,x))])
```

```python
# Extract unique elements from a list using list comprehension (without using set()).
a=[1,2,2,3,1,4,5,5]
b=[]
[ b.append(i) for i in a if i not in b]
print(b)
```

```python
a=[1,2,2,3,1,4,5,5]
print([i for i in a if a.count(i)==1])
```

```python
# Generate a list of reversed strings only for strings longer than 4 characters.
# Example: ['apple', 'bat', 'banana', 'dog'] → ['elppa', 'ananab']

print([i[::-1] for i in ['apple', 'bat', 'banana', 'dog'] if len(i)>4])
```

```python
# Create a list of numbers that are palindromes between 1 and 200.
print([i for i in range(1,200) if str(i)==str(i)[::-1]])
```

```python
# Create a dictionary that maps numbers from 1 to 5 to their squares.
# Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

print({ i:i**2 for i in range(1,6)})
```

```python
# Given a string, count the frequency of each character using dictionary comprehension.
a='balavenu'
print({i:a.count(i) for i in a})
```

```python
# Convert two lists into a dictionary using dictionary comprehension.
# keys = ['a', 'b', 'c'], values = [1, 2, 3] → {'a': 1, 'b': 2, 'c': 3}

keys=['a','b','c']
values=[1,2,3]
print({keys[i]:values[i] for i in range(len(keys))})
```

```python
# Create a dictionary that maps each word in a sentence to its length.
a='bala venu'
print({i:len(i) for i in a.split()})
```

```python
# Given a dictionary of items and prices, create a new dictionary with prices increased by 10%.
a={'banana':100,'apple':200,'carrot':350}
print({i:j+int(((10/100)*j)) for i,j in a.items()})
```

```python
# Create a dictionary that swaps keys and values of an existing dictionary.
a={1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
print({j:i for i,j in a.items()})
```

```python
# Filter out dictionary items where the value is less than 50.
# {'a': 10, 'b': 60, 'c': 30, 'd': 80} → {'b': 60, 'd': 80}
a={'a': 10, 'b': 60, 'c': 30, 'd': 80}
print({i:j for i,j in a.items() if j>50})
```

```python
# Create a dictionary where keys are numbers from 1 to 10 and values are 'even' or 'odd'.
print({i:('even' if i%2==0 else 'odd') for i in range(1,11)})
```

```python
# Merge two dictionaries using dictionary comprehension.
a={1: 1, 4: 2, 9: 3, 16: 4, 25: 5}
b={'a': 10, 'b': 60, 'c': 30, 'd': 80}
{a.update({i:j}) for i,j in b.items()}
print(a)
```

```python
# Given a dictionary of students and their marks, create a new dictionary containing only students who scored above 75.
a={'bala':76,'venu':70,'balu':80}
print({i:j for i,j in a.items() if j>75})
```

```python
# Create a nested dictionary for y in range(1, 3) for x in range(1, 3).
print({x: {y: x*y for y in range(1, 3)} for x in range(1,3)})
```

```python
# Given a dictionary of words and their frequencies, normalize the frequencies so they sum to 1.
a={'banana':20,'apple':10,'carrot':15,'mango':5}
tot=sum(a.values())
print({i:(j/tot) for i,j in a.items()})
```
