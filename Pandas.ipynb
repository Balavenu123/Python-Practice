```python
import numpy as np
import pandas as pd
print(dir(pd))
print(len(dir(pd)))
```

```python
# Series is used to create one column representation
a=np.array([10,20,30,40,50])
b=pd.Series(a)
b
```

```python
# DataFrame to create multiple column data
a={'name':['bala','venu','balineni','kamal sir','ramu','raghu'],
   'age':[24,23,25,34,43,23],
   'loc':['bang','hyd','delhi','chennai','mumbai','karnataka'],
   'experiance':[3,4,2,9,4,6],
   'salary':[1200,3000,7000,9999,3004,9889]}
b=pd.DataFrame(a)
b
```

```python
b['name'][3]
```

```python
# loc is used to access more values
b.loc[1:4,'loc':'salary']
```

```python
# iloc is used to access more values
b.iloc[1:3,1:5]
```

```python
# renam is used to change the column name
b=b.rename(columns={'loc':'Area'})
b
```

```python
 # axis=0 --> means rows
 # axis=1 --> means col
 # drop is used to delet the data(column)
c=b.drop(['experiance'],axis=1)
c
```

```python
# adding new col and values
b['emi']=[10,20,30,40,50,60]
b
```

```python
# isnull() is used to find there is any null value
b.isnull()
```

```python
# isnull().sum() is used to find total no of null values
b.isnull().sum()
```

```python
# dropna is used to delete null value rows
b.dropna()
```

```python
# fillna() is used to fill null values with our own values
b['tax']=[np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
b['tax']=b['tax'].fillna(20)
b
```

```python
# to_csv convert data into excel
b.to_csv('emp_data.csv',index=False)
```
