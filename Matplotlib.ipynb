```python
import matplotlib.pyplot as plt
```

```python
india=[1,3,5,7,6]
usa=[3,4,5,2,1]
plt.figure(figsize=(5,3))
plt.title('Scatter plot')
plt.xlabel('Indian_values')
plt.ylabel('Usa_values')
plt.scatter(x=india,y=usa,color='g',marker='s')
plt.show
```

```python
india=[1,3,5,7,6]
usa=[3,4,5,2,1]
plt.figure(figsize=(5,3))
plt.title('Line_plot')
plt.xlabel('Indian_values')
plt.ylabel('Usa_values')
plt.plot(india,usa,color='g',marker='s')
plt.show
```

```python
india=[1,3,5,7]
usa=[3,4,5,2]
uk=[3,7,0,4]
plt.figure(figsize=(5,3))
plt.title('Line_plot')
plt.xlabel('Indian_values')
plt.ylabel('Usa_values')
plt.plot(india,usa,label='india_usa')
plt.plot(india,uk,label='india_uk')
plt.legend(loc=0)
plt.show
```

```python
india=[1,3,5,7]
usa=[3,4,5,2]
uk=[3,7,0,4]
plt.figure(figsize=(5,3))
plt.subplot(1,2,1)
plt.title('India_Usa')
plt.plot(india,usa,marker='*',color='r')
plt.subplot(1,2,2)
plt.title('India_Uk')
plt.plot(india,uk,marker='*',color='r')
plt.show
```

```python
india=[1,3,5,7]
usa=[3,4,5,2]
uk=[3,7,0,4]
plt.figure(figsize=(5,3))
plt.subplot(2,1,1)
plt.title('India_Usa')
plt.plot(india,usa,marker='*',color='r')
plt.subplot(2,1,2)
plt.title('India_Uk')
plt.plot(india,uk,marker='*',color='r')
plt.show
```

```python
import numpy as np
```

```python
data=np.random.randn(1000)
plt.figure(figsize=(5,3))
plt.hist(x=data,bins=5)
plt.show()
```

```python
products = ['Product X', 'Product Y', 'Product Z']
revenue_q1 = [50, 70, 60]
plt.figure(figsize=(5,3))
plt.bar(products,revenue_q1,color=['r','g','y'])
plt.show()
```

```python
products = ['Product X', 'Product Y', 'Product Z']
revenue_q1 = [50, 70, 60]
plt.figure(figsize=(5,3))
plt.barh(products,revenue_q1,color=['r','g','y'])
plt.show()
```

```python
products = ['Product X', 'Product Y', 'Product Z']
revenue_q1 = [50, 70, 60]
plt.figure(figsize=(5,3))
plt.pie(revenue_q1,labels=products,shadow=True,autopct='%1.4f%%')
plt.show()
```
