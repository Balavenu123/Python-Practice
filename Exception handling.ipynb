```python
import sys
try:
  a=int(input())
  b=int(input())
  c=a/b
  print(c)
except Exception as e:
  er_mss,er_type,er_no= sys.exc_info()
  print(f'we got {er_mss} because of {er_type} at line no {er_no.tb_lineno}')
```
