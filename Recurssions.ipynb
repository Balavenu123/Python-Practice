```python
# Write a recursive function factorial(n) that returns the factorial of n.

def fact(a):
  if a==1 or a==0:
    return 1
  else:
    return a*fact(a-1)
sol=fact(5)
print(sol)
```

```python
# Write a function sum_n(n) that returns the sum of all numbers from 1 to n

def sum_n(n):
  if n==0:
    return 0
  else:
    return n+sum_n(n-1)
sol=sum_n(4)
print(sol)
```

```python
# Recursively print all numbers from 1 to n.
def num(n):
  if n==0:
    return
  num(n-1)
  print(n,end=' ')
num(4)
```

```python
# Recursively print numbers from n down to 1.
def num(n):
  if n==0:
    return
  print(n,end=' ')
  num(n-1)
num(4)
```

```python
# Write a recursive function to find the nth Fibonacci number.

def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)
sol=fib(7)
print(sol)
```

```python
# Write a function sum_digits(n) that returns the sum of all digits in a given number.
def sum_digits(n):
  if n==0:
    return 0
  else:
    return (n%10)+sum_digits(n//10)
sol=sum_digits(1234)
print(sol)
```

```python
# Recursively reverse a string without using slicing.

def reverse(str):
    if len(str) == 0:
        return str
    return reverse(str[1:]) + str[0]
sol=reverse('bala')
print(sol)
```
