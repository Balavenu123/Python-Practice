```python
# Read the contents of file1.txt and write them into a new file file2.txt.
a=open('notes.txt','r')
b=a.read()
c=open('venu.txt','a')
c.write(b)
c.close()
a.close()
```

```python
# Read a file and find the longest word in it.
a=open('notes.txt','r')
b=a.read()
c=b.split()
large=''
for i in c:
  if len(i)>len(large):
    large=i
print(large)
a.close()
```
