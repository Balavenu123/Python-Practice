```python
import sys
```

```python
class PERFECT:
  try:
      def __init__(self,a):
        self.a=a
      def perfect(self):
        self.tot=0
        for i in range (1,self.a):
          if self.a%i==0:
            self.tot=self.tot+i
        if self.a==self.tot:
          print('perfect')
        else:
          print('not perfect')
  except Exception as e:
    ex_type,ex_msg,ex_no= sys.exc_info()
    print(f'Error in line no:{ex_no.tb_lineno} due to:{ex_type}')
bala=PERFECT(28)
bala.perfect()
```

```python
class PRIME:
  try:
    def __init__(self,a,b):
      self.a=a
      self.b=b
    def prime_no(self):
      for i in range(self.a,self.b):
        if i>1:
          for j in range(2,i):
            if i%j==0:
              break;
          else:
             print(i)

  except Exception as e:
    ex_type,ex_msg,ex_no= sys.exc_info()
    print(f'Error in line no:{ex_no.tb_lineno} due to:{ex_type}')
venu=PRIME(1,100)
venu.prime_no()
```

```python
a='aaaabbbccccaaggg'
s=""
for i in range(len(a)):
  b=a[i]
  if i==0:
    c=0
    for j in a:
      if j==b:
        c += 1
      else:
        break
    s += b+str(c)
  else:
    d=a[i:]
    if b==a[i-1]:
      continue
    else:
      c=0
      for k in d:
        if k==b:
          c += 1
        else:
          break
      s += b+str(c)
print(s)
```

```python
# Create a Rectangle class with length and width set through the constructor; add a method to calculate area.

class RECTANGLE:
  def __init__(self,length,breadth):
    self.length=length
    self.breadth=breadth
  def area(self):
    tot_area=self.length*self.breadth
    print(f'Area of rectangle is {tot_area}')
bala=RECTANGLE(4,5)
bala.area()
```
