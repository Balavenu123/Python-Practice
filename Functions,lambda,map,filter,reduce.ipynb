```python
# Define a function max_of_three(a, b, c) that returns the largest of three numbers.

def max_of_three(a,b,c):
  if a>b and a>c:
    return a
  elif b>a and b>c:
    return b
  else:
    return c
a,b,c=eval(input())
sol=max_of_three(a,b,c)
print(f'{sol} is big')
```

```python
# Write a function factorial_iterative(n) that calculates the factorial of a non-negative integer n using a loop.

def factorial_iterative(n):
  if n>0:
    fact=1
    while n!=0:
      fact=fact*n
      n=n-1
    return fact

n=int(input())
sol=factorial_iterative(n)
print(f'{sol} is factorial of {n}')
```

```python
# Create a function count_vowels(s) that takes a string and returns the number of vowels it contains (case-insensitive).

def count_vowels(s):
  count=0
  for i in s:
    if i in 'aeiouAEIOU':
      count +=1
  return count

s=input()
sol=count_vowels(s)
print(f'there are {sol} vowels in {s} ')
```

```python
# Write a function is_prime(n) that takes an integer n and returns True if it's a prime number, and False otherwise.

def is_prime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
    return True
n=int(input())
sol=is_prime(n)
if sol==True:
  print(f'{n} is prime number')
else:
  print(f'{n} is not prime')
```

```python
# Create a function is_palindrome_pro(s) that checks if a string is a palindrome, ignoring spaces, punctuation, and letter casing.

def is_palindrome(s):
  a=s.lower()
  b=''
  for i in a:
    if i.isalnum():
      b=b+i
    else:
      pass
  return b

s=input()
sol=is_palindrome(s)
if sol==sol[::-1]:
  print(f'{s} is palindrome')
else:
  print(f'{s} is not palindrome')
```

```python
# Rewrite the factorial function using recursion.

def fact(n):
  if n==1 or n==0:
    return 1
  else:
    return n*fact(n-1)

sol=fact(5)
print(sol)
```

```python
# Create a function find_duplicates(lst) that takes a list and returns a new list containing only the unique duplicate elements.

def unique(a):
  b=[]
  for i in a:
    if i not in b:
      b.append(i)
  return b

a=eval(input())
sol=unique(a)
print(sol)
```

```python
# Write a function gcd(a, b) that computes the greatest common divisor of two positive integers using the Euclidean algorithm.

def gcd(a, b):
  while b:
    a, b = b, a % b
  return a

print(f"The GCD of 54 and 24 is: {gcd(54, 24)}")
```

```python
# Write a recursive function sum_digits(n) that calculates the sum of the digits of a non-negative integer.


def sum_digits(n):
  if n<10:
    return n
  else:
    return n%10+sum_digits(n//10)
sol=sum_digits(123)
print(sol)
```
