```python
# Write a Python function that uses a for loop to combine odd-indexed elements from one list and even-indexed elements from another into a third list.

def lists(a,b):
  c=[]
  for i in a:
    if a.index(i)%2==0:
      c.append(i)
  for j in b:
    if b.index(j)%2!=0:
      c.append(j)
  return(c)
a=[1,2,3,4,5]
b=[6,7,8,9,0]
sol=lists(a,b)
print(sol)
```

```python
# Given a dictionary grades = {"Alice": 95, "Bob": 88}, use a for loop to print each key.

grades = {"Alice": 95, "Bob": 88}
for i in grades:
  print(i)
```

```python
# Use a for loop to iterate through a dictionary and print each key-value pair.

grades = {"Alice": 95, "Bob": 88}
for i in grades:
  print(i,grades[i])
for i,j in grades.items():
  print(i)
  print(j)
```

```python
# Write a program that uses an if/elif/else statement inside a for loop to print a message based on the value in a dictionary (e.g., for grades, print "Excellent" for scores >= 90).

score={'bala':98,'venu':78,'hh':50}
mess=''
for i,j in score.items():
  if j>90:
    mess='Excellent'
  elif j<90 and j>50:
    mess='good'
  else:
    mess='avg'
  print(f'{i} got {mess} marks')
```

```python
# Write a program that uses a while loop to repeatedly ask the user for a key-value pair and add it to a dictionary. Stop when the user types "quit".

a={}
while True:
    key=input("Enter a key (or 'quit'): ")
    if key.lower() == 'quit':
        break
    value=input(f"Enter the value for '{key}': ")
    a[key] = value
print(a)
```

```python
# rotating a list with one place value
a=[1,2,3,4,5]
b=a[0]
for i in range(1,len(a)):
  a[i-1]=a[i]
a[len(a)-1]=b
print(a)
```

```python
# Given a dictionary, use a for loop to swap its keys and values, creating a new dictionary.
a={'bala':100, 'venu': 200, 'chintu': 300}
b={}
c=[]
d=[]
for i,j in a.items():
  c.append(i)
  d.append(j)
for i in range(0,len(c)):
  b[d[i]]=c[i]
print(b)
```

```python
# Write a Python program that uses a for loop to sort a dictionary by its values.
a={'a':13,'b':10,'c':89,'d':0}
sorted_a = {}

for key in sorted(a, key=a.get):
    sorted_a[key] = a[key]
print(f"Sorted dictionary: {sorted_a}")
```
